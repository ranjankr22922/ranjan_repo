```python
from matplotlib import pyplot as plt
import numpy as np
```

Legend Plot

```python
x1=[1,5,8,3]
x2=[2,1,6,7]
y1=[4,5,1,3]
y2=[2,0,8,5]
plt.plot(x1,x2,'r:*',markerfacecolor='blue', markeredgecolor='blue', label='first')
plt.plot(y1,y2,'g-*',markerfacecolor='yellow', markeredgecolor='yellow', label='second')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plotting two sets of data')
plt.legend()
plt.show()
```

Pie Chart

```python
cat = ['Action', 'Draw', 'Thriller', 'Comedy']
vote = [3,4,8,7]
plt.pie(vote)
plt.show()
```

```python
plt.title("Movie Votecast")
plt.pie(vote, labels=cat, labeldistance=1.1)
plt.legend(loc=4)
plt.show()
```

Pulling out Wedges(explode)

```python
plt.pie(vote, labels=cat, labeldistance=1.1, explode=[0,0.1,0,0.1])
plt.legend(loc=3)
plt.show()
```

Change Angle (startangle)

```python
plt.pie(vote, labels=cat, labeldistance=1.1, startangle=90)
plt.legend(loc=4)
plt.show()
```

Setting Colors of each Wedge

```python
plt.pie(vote, labels=cat, labeldistance=1.1, colors=['red','g','b','y'])
plt.legend(loc=4)
plt.show()
```

Display Pecentage and Change distance inside each wedge


```python
plt.pie(vote, labels=cat, labeldistance=1.1, shadow=True, autopct='%0.2f%%', pctdistance=0.6)
plt.legend(loc=4)
plt.show()
```

Change the Radius 

```python
plt.title("Movie Votecast")
plt.pie(vote, labels=cat, labeldistance=1.1, radius=1.5)
plt.legend(loc=4)
plt.show()
```

Set Label Size 

```python
plt.title("Movie Votecast")
plt.pie(vote, labels=cat, labeldistance=1.1, textprops={'fontsize':13})     #text properties
plt.legend(loc=4)
plt.show()
```

Setting Wedges Border

```python
plt.title("Movie Votecast")
plt.pie(vote, labels=cat, labeldistance=1.1, shadow=True, wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':':'})
plt.legend(loc=4)
plt.show()
```
